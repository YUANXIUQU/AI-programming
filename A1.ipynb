```python
import numpy as np
#ANN using ReLU function
#Initialize parameters
#weights
w1_3 = 0.5
w2_3 = 0.4
w1_4 = 0.9
w2_4 = 1.0
w1_5 = 0.7
w2_5 = 0.6

w3_6 = -1.2
w4_6 = 1.1
w5_6 = 1.3

#Biasis
Bias3 = 0.4
Bias4 = -0.1
Bias5 = 0.3
Bias6 = -0.4

#learning rate
Alpha = 0.01
Niteration = 50000

def relu(X):
  y = np.maximum(0, X)
  return y
def derivative_ReLU(X):
  return (1 if X > 0 else 0)

# Training data
x1 = [1, 0, 1, 0]
x2 = [1, 1, 0, 0]
Yd6 = [0, 1, 1, 0]

p = 1
# Start training
while p < Niteration:
    EpocSumError = 0

    for i in range(4):
        print(f"Iteration: {p}, Case: {i + 1}")

        # Feedforward
        X3 = x1[i] * w1_3 + x2[i] * w2_3 + Bias3
        y3 = relu(X3)  # ReLU activation

        X4 = x1[i] * w1_4 + x2[i] * w2_4 + Bias4
        y4 = relu(X4)  # ReLU activation

        X5 = x1[i] * w1_5 + x2[i] * w2_5 + Bias5
        y5 = relu(X5)  # ReLU activation

        X6 = y3 * w3_6 + y4 * w4_6 + y5 * w5_6 + Bias6
        y6 = relu(X6)  # ReLU activation

        # Backpropagation
        e6 = Yd6[i] - y6  # Error at output

        D6 = derivative_ReLU(X6) * e6  # Delta for output layer

        Wcurrent3_6 = w3_6
        Wcurrent4_6 = w4_6
        Wcurrent5_6 = w5_6

        # Update weights and biases for neuron 6
        w3_6 += Alpha * y3 * D6
        w4_6 += Alpha * y4 * D6
        w5_6 += Alpha * y5 * D6
        Bias6 += Alpha * D6

        # Calculate delta for neuron 3
        D3 = derivative_ReLU(X3) * D6 * Wcurrent3_6
        w1_3 += Alpha * x1[i] * D3
        w2_3 += Alpha * x2[i] * D3
        Bias3 += Alpha * D3

        # Calculate delta for neuron 4
        D4 = derivative_ReLU(X4) * D6 * Wcurrent4_6
        w1_4 += Alpha * x1[i] * D4
        w2_4 += Alpha * x2[i] * D4
        Bias4 += Alpha * D4

        # Calculate delta for neuron 5
        D5 = derivative_ReLU(X5) * D6 * Wcurrent5_6
        w1_5 += Alpha * x1[i] * D5
        w2_5 += Alpha * x2[i] * D5
        Bias5 += Alpha * D5

        EpocSumError += e6 ** 2  # Sum of squared errors

    print("Epoch error:", EpocSumError)

    if EpocSumError < 0.001:
        break

    p += 1

# Final output
print("Training completed after", p, "iterations")
print(f"w1_3: {w1_3}, w2_3: {w2_3}, Bias3: {Bias3}")
print(f"w1_4: {w1_4}, w2_4: {w2_4}, Bias4: {Bias4}")
print(f"w1_5: {w1_5}, w2_4: {w2_5}, Bias5: {Bias5}")
print(f"w3_5: {w3_6}, w4_5: {w4_6}, w5_6: {w5_6}, Bias6: {Bias6}")

#Test
for i in range(4):

  X3 = x1[i] * w1_3 + x2[i] * w2_3 + Bias3
  y3 = max(0,X3)  # ReLU activation

  X4 = x1[i] * w1_4 + x2[i] * w2_4 + Bias4
  y4 = max(0,X4)  # ReLU activation

  X5 = x1[i] * w1_5 + x2[i] * w2_5 + Bias5
  y5 = max(0,X5)  # ReLU activation

  X6 = y3 * w3_6 + y4 * w4_6 + y5 * w5_6 + Bias6
  y6 = max(0,X6)
  print (f"result{i+1} x1: {x1[i]}, x2: {x2[i]} output: {y6}")
```
